```
def calculo(dataframe, coluna): # função para criar uma lista com os valores dados extraídos do Dataframe original

  # "to_datetime" para converter as datas em objetos de datetime, assim podendo ser manipulados. " errors='coerce' "apenas para converter valores não analisáveis
  dataframe['data_inversa'] = pd.to_datetime(dataframe['data_inversa'], errors='coerce') 
  dataframe['mes'] = pd.DatetimeIndex(dataframe['data_inversa']).month # Criando uma nova coluna chamada "mes" no dataframe original, contendo o mês extraído da coluna "data_inversa"

  lista = [] # lista para armazenar os valores da contagem

  for i in range (1, 13): # percorrendo os meses 
    lista.append(dataframe[(dataframe['mes'] == i) & (dataframe[coluna] > 0)].shape[0]) # analisando quantos valores em uma determinada coluna de um determinado mês existem, e colocando numa lista

  return lista 

def proporcao (lista_ilesos, lista_envolvidos): # função para criar uma lista com o cálculo da proporcao de ilesos e envolvidos

  proporcao = []

  for i in range(12):
    if lista_ilesos[i] != 0 and lista_envolvidos[i] != 0: # condição para evitar erro de divisão por zero
      proporcao.append((lista_ilesos[i] / lista_envolvidos[i])*100) # calculando a proporção e multiplicando por 100, para ter a porcentagem
    else : proporcao.append('-') # se valores forem diferentes de zero (nesse caso eles só poderiam ser NaN), serão colocados '-' em seu lugar
  
  return proporcao

# Criando Series a partir da função "calculo", que cria uma lista com os valores do dataframe de 2019 ou 2020
# O Dataframe e a coluna a ser analisada são passados como parâmetros
# A função analisa quantos valores há em cada mês e preenche os índices dessa forma, criando uma lista com 11 índices (De 0 à 11).
# Após isso a lista é transformada em uma Serie
feridosGraves2019 = pd.Series(calculo(df2019, 'feridos_graves')) 
feridosGraves2020 = pd.Series(calculo(df2020, 'feridos_graves')) 
pessoasIlesas2019 = pd.Series(calculo(df2019, 'ilesos'))
pessoasIlesas2020 = pd.Series(calculo(df2020, 'ilesos'))
envolvidos2019 = pd.Series(calculo(df2019, 'pessoas'))
envolvidos2020 = pd.Series(calculo(df2019, 'pessoas'))

# Usando a função "proporcao" para criar uma serie com dados calculados relacionando duas listas (Ilesos e Pessoas envolvidas)
proporcao2019 = pd.Series( proporcao( calculo(df2019, 'ilesos'),  calculo(df2019, 'pessoas')))
proporcao2020 = pd.Series( proporcao( calculo(df2020, 'ilesos'),  calculo(df2020, 'pessoas')))

# Usando a função ".concat" para criar um Dataframe com todas as informações das Series criadas anteriormente
All = pd.concat([feridosGraves2019, feridosGraves2020, pessoasIlesas2019, pessoasIlesas2020, proporcao2019, proporcao2020], axis = 1)
# Criando os novos índices e colocando no dataframe com a função ".index"
All.index = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
# Criando as novas colunas e colocando no dataframe com a função ".columns"
All.columns = ['Feridos_graves_2019','Feridos_graves_2020','Ilesas_2019','Ilesas_2020','Proporção2019 (%)','Proporção2020 (%)']

All
```

(c) mostre, reorganizando o conjunto de dados, se a pandemia conseguiu diminuir ou não a incidência de acidentes nas rodovias federais.

```
# FEITO COM ESTRATÉGIAS DA QUESTÃO ANTERIOR

def calculo(dataframe, coluna): # função para criar uma lista com os valores dados extraídos do Dataframe original

  # "to_datetime" para converter as datas em objetos de datetime, assim podendo ser manipulados. " errors='coerce' "apenas para converter valores não analisáveis
  dataframe['data_inversa'] = pd.to_datetime(dataframe['data_inversa'], errors='coerce') 
  dataframe['mes'] = pd.DatetimeIndex(dataframe['data_inversa']).month # Criando uma nova coluna chamada "mes" no dataframe original, contendo o mês extraído da coluna "data_inversa"

  lista = [] # lista para armazenar os valores da contagem

  for i in range (1, 13): # percorrendo os meses 
    lista.append(dataframe[(dataframe['mes'] == i) & (dataframe[coluna] > 0)].shape[0]) # analisando quantos valores em uma determinada coluna de um determinado mês existem, e colocando numa lista

  return lista 

def diferenca (Acidentes2019, Acidentes2020): # função para criar uma lista com o cálculo da diferenca de acidentes em 2019 e 2020

  diferenca = []

  for i in range(12):
    if Acidentes2019[i] != 0 and Acidentes2020[i] != 0: # condição para evitar erro de divisão por zero

      if (Acidentes2019[i] / Acidentes2020[i]) < 1:  diferenca.append("Aumentou") # Condição para saber se houve aumento ou diminuição nos casos 
      else: diferenca.append("Diminuiu")

    else : diferenca.append('-') # se valores forem diferentes de zero (nesse caso eles só poderiam ser NaN), serão colocados '-' em seu lugar
  
  return diferenca

# Criando Series a partir da função "calculo", que cria uma lista com os valores do dataframe de 2019 ou 2020
# O Dataframe e a coluna a ser analisada são passados como parâmetros
# A função analisa quantos valores há em cada mês e preenche os índices dessa forma, criando uma lista com 11 índices (De 0 à 11).
# Após isso a lista é transformada em uma Serie
QuantAcidentes2019 = pd.Series(calculo(df2019, 'id'))  # Como todas as linhas da coluna 'id' são maiores que 0, ela foi utilizada para identificar quantos acidentes tinham em cada mês na função de calculo
QuantAcidentes2020 = pd.Series(calculo(df2020, 'id'))

# Usando a função "diferenca" para criar uma serie com dados calculados relacionando duas listas (Quantidades de acidentes em 2019 e 2020)
diferencaAnos = pd.Series(diferenca(calculo(df2019, 'id'), calculo(df2020, 'id')))

# Usando a função ".concat" para criar um Dataframe com todas as informações das Series criadas anteriormente
All = pd.concat([QuantAcidentes2019, QuantAcidentes2020, diferencaAnos], axis = 1)
# Criando os novos índices e colocando no dataframe com a função ".index"
All.index = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
# Criando as novas colunas e colocando no dataframe com a função ".columns"
All.columns = ['Quant_Acidentes_2019','Quant_Acidentes_2020', ' Diferença ']

All
```
